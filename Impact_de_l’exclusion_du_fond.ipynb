```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
import matplotlib.pyplot as plt
import scipy.io as sio
import os
from tqdm import tqdm
from sklearn.utils.class_weight import compute_class_weight

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Define a simpler and more proven 3D-CNN architecture for hyperspectral imaging
import torch
import torch.nn as nn
import torch.nn.functional as F

import torch
import torch.nn as nn
import torch.nn.functional as F

class HybridSN(nn.Module):
    def __init__(self, in_channels, patch_size, n_classes):
        super(HybridSN, self).__init__()

        self.patch_size = patch_size

        # 3D Convolution layer
        self.conv3d_1 = nn.Conv3d(1, 8, kernel_size=(3, 3, 3), stride=(1, 1, 1), padding=(1, 1, 1))
        self.bn3d_1 = nn.BatchNorm3d(8)

        # Determine spatial dimensions after 3D operations
        dummy_input = torch.zeros(1, 1, in_channels, patch_size, patch_size)
        x = F.relu(self.bn3d_1(self.conv3d_1(dummy_input)))

        # Reshape for 2D convolution (N,C,D,H,W) -> (N,C*D,H,W)
        spatial_dim = patch_size
        n_dim = x.size(1) * x.size(2)  # Channels * Depth
        x = x.view(x.size(0), n_dim, spatial_dim, spatial_dim)

        # 2D Convolution layer
        self.conv2d_1 = nn.Conv2d(n_dim, 64, kernel_size=3, stride=1, padding=1)
        self.bn2d_1 = nn.BatchNorm2d(64)

        # Global average pooling
        x = F.relu(self.bn2d_1(self.conv2d_1(x)))
        x = F.adaptive_avg_pool2d(x, (1, 1))
        fc_size = x.view(x.size(0), -1).shape[1]

        # Fully connected layer
        self.fc1 = nn.Linear(fc_size, n_classes)

        print(f"Model initialized with FC input size: {fc_size}")

    def forward(self, x):
        # 3D Convolution path
        x = F.relu(self.bn3d_1(self.conv3d_1(x)))

        # Reshape for 2D convolution (N,C,D,H,W) -> (N,C*D,H,W)
        x = x.view(x.size(0), -1, x.size(3), x.size(4))

        # 2D Convolution path
        x = F.relu(self.bn2d_1(self.conv2d_1(x)))

        # Global average pooling
        x = F.adaptive_avg_pool2d(x, (1, 1))
        x = x.view(x.size(0), -1)

        # Fully connected path
        x = self.fc1(x)

        return x

# Custom Dataset for Indian Pines with data augmentation
class IndianPinesDataset(Dataset):
    def __init__(self, data, labels, patch_size=11, transforms=None, use_augmentation=False):
        self.data = data
        self.labels = labels
        self.patch_size = patch_size
        self.transforms = transforms
        self.use_augmentation = use_augmentation

        # Dimensions
        self.bands, self.height, self.width = data.shape

        # Create valid indices (excluding boundary pixels)
        self.valid_indices = []
        pad = patch_size // 2
        for i in range(pad, self.height - pad):
            for j in range(pad, self.width - pad):
                  self.valid_indices.append((i, j))

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        # Get center pixel coordinates
        i, j = self.valid_indices[idx]
        pad = self.patch_size // 2

        # Extract patch
        patch = self.data[:, i-pad:i+pad+1, j-pad:j+pad+1].copy()

        # Get label
        label = self.labels[i, j]

        # Apply data augmentation
        if self.use_augmentation and np.random.random() > 0.5:
            # Random flip
            if np.random.random() > 0.5:
                patch = np.flip(patch, axis=1).copy()
            if np.random.random() > 0.5:
                patch = np.flip(patch, axis=2).copy()

            # Random rotation (90, 180, 270 degrees)
            k = np.random.randint(1, 4)
            patch = np.rot90(patch, k=k, axes=(1, 2)).copy()

            # Small random noise
            if np.random.random() > 0.7:
                noise = np.random.normal(0, 0.01, patch.shape)
                patch = patch + noise

        # Convert to tensor
        patch = torch.FloatTensor(patch).unsqueeze(0)  # Add channel dimension for 3D convolution

        if self.transforms:
            patch = self.transforms(patch)

        return patch, label

# Function to load and preprocess the data
def load_indian_pines_data(data_path='Indian_pines.mat', gt_path='Indian_pines_gt.mat'):
    # Load data
    data = sio.loadmat(data_path)['indian_pines']
    gt = sio.loadmat(gt_path)['indian_pines_gt']

    # Normalize data using MinMaxScaler (better for preserving spatial info)
    scaler = MinMaxScaler()
    h, w, b = data.shape
    data_reshaped = data.reshape((h*w, b))
    data_normalized = scaler.fit_transform(data_reshaped)
    data = data_normalized.reshape((h, w, b))

    # Transpose to bands-first format (bands, height, width)
    data = np.transpose(data, (2, 0, 1))

    return data, gt

from sklearn.metrics import accuracy_score, f1_score
import torch
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt

def train_model(model, train_loader, criterion, optimizer, device, num_epochs=100):
    train_losses = []
    val_losses = []
    val_accuracies = []
    val_f1_scores = []

    best_val_acc = 0.0
    best_val_f1 = 0.0
    best_model_state = None


    # Early stopping
    patience = 15
    counter = 0

    for epoch in range(num_epochs):
        # Training
        model.train()
        running_loss = 0.0

        # Use tqdm for progress bar
        loop = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')

        for inputs, labels in loop:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)

            loss.backward()
            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

            # Update progress bar
            loop.set_postfix(loss=loss.item())

        epoch_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_loss)

        # Validation (using the same data as training)
        model.eval()
        val_loss = 0.0
        val_preds = []
        val_true = []

        with torch.no_grad():
            for inputs, labels in train_loader:  # Using the same train_loader for validation
                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)

                val_loss += loss.item() * inputs.size(0)

                _, preds = torch.max(outputs, 1)
                val_preds.extend(preds.cpu().numpy())
                val_true.extend(labels.cpu().numpy())

        val_epoch_loss = val_loss / len(train_loader.dataset)
        val_losses.append(val_epoch_loss)

        val_acc = accuracy_score(val_true, val_preds)
        val_accuracies.append(val_acc)

        # Calculate F1 score for more comprehensive evaluation
        val_f1 = f1_score(val_true, val_preds, average='weighted')
        val_f1_scores.append(val_f1)

        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_loss:.4f}, '
              f'Val Loss: {val_epoch_loss:.4f}, '
              f'Val Acc: {val_acc:.4f}, '
              f'Val F1: {val_f1:.4f}')

        # Save best model state (in memory)
        if val_f1 > best_val_f1:
            best_val_f1 = val_f1
            best_val_acc = val_acc
            best_model_state = model.state_dict().copy()
            counter = 0
            print(f"New best model saved with F1 Score: {val_f1:.4f}")
        else:
            counter += 1

        # Early stopping
        if counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

    # Load best model state back into the model
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
        print(f"Loaded best model with F1 Score: {best_val_f1:.4f}")

    # Plot training curves
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(val_accuracies, label='Val Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(val_f1_scores, label='Val F1 Score')
    plt.xlabel('Epoch')
    plt.ylabel('F1 Score')
    plt.legend()

    plt.tight_layout()
    plt.savefig(f'training_curves_hybridsn_patch{model.patch_size}.png')

    print(f"Training complete. Best validation accuracy: {best_val_acc:.4f}, Best F1 Score: {best_val_f1:.4f}")

    return model, train_losses
# Function to predict on the entire image with CRF post-processing
def predict_image(model, data, patch_size, num_classes, device):
    model.eval()

    # Get dimensions
    bands, height, width = data.shape
    pad = patch_size // 2

    # Initialize prediction map
    pred_map = np.zeros((height, width))

    # Pad the data
    padded_data = np.pad(data, ((0, 0), (pad, pad), (pad, pad)), mode='reflect')

    # Process in batches for efficiency
    batch_size = 32  # Increased batch size for faster prediction
    coords = [(i, j) for i in range(height) for j in range(width)]
    num_batches = (len(coords) + batch_size - 1) // batch_size

    # Store class probabilities for all pixels
    prob_map = np.zeros((height, width, num_classes))

    with torch.no_grad():
        for batch_idx in tqdm(range(num_batches), desc="Predicting"):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(coords))
            batch_coords = coords[start_idx:end_idx]

            batch_patches = []
            for i, j in batch_coords:
                # Extract patch
                patch = padded_data[:, i:i+patch_size, j:j+patch_size]
                batch_patches.append(patch)

            # Stack patches and convert to tensor
            batch_patches = torch.FloatTensor(np.array(batch_patches)).unsqueeze(1)  # Add channel dimension

            # Predict
            outputs = model(batch_patches.to(device))
            probs = torch.softmax(outputs, dim=1).cpu().numpy()
            _, preds = torch.max(outputs, 1)

            # Store predictions and probabilities
            for idx, (i, j) in enumerate(batch_coords):
                pred_map[i, j] = preds[idx].item() + 1  # Convert back to 1-based indexing
                prob_map[i, j] = probs[idx]

    # Try to use CRF for post-processing if pydensecrf is available
    try:
        import pydensecrf.densecrf as dcrf
        from pydensecrf.utils import unary_from_softmax

        print("Applying CRF post-processing...")

        # Create CRF model
        d = dcrf.DenseCRF2D(width, height, num_classes)

        # Reshape probability map for CRF
        U = unary_from_softmax(np.transpose(prob_map, (2, 0, 1)))
        d.setUnaryEnergy(U)

        # Add pairwise potentials
        d.addPairwiseGaussian(sxy=3, compat=3)
        d.addPairwiseBilateral(sxy=80, srgb=13, rgbim=np.transpose(data[:3], (1, 2, 0)), compat=10)

        # Perform inference
        Q = d.inference(5)
        pred_map_crf = np.argmax(np.array(Q).reshape((num_classes, height, width)), axis=0) + 1

        print("CRF post-processing complete.")

    except ImportError:
        print("pydensecrf not available, using median filter instead.")
        # Apply median filter as fallback
        from scipy import ndimage
        pred_map_crf = ndimage.median_filter(pred_map, size=3)

    return pred_map, pred_map_crf, prob_map

# Main execution function
def run_indian_pines_segmentation(patch_size=11):
    # Check if patch size is even, and if so, make it odd
    if patch_size % 2 == 0:
        print(f"Warning: Patch size {patch_size} is even. Using {patch_size+1} instead.")
        patch_size = patch_size + 1

    # Check for GPU
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'Using device: {device}')

    # Load data
    print("Loading and preprocessing data...")
    data, gt = load_indian_pines_data()

    # Dataset parameters
    num_classes = 17  # Indian Pines has 16 classes
    batch_size = 32

    print(f"Using patch size: {patch_size}")

    # Create dataset with data augmentation for training
    train_dataset = IndianPinesDataset(data, gt, patch_size=patch_size, use_augmentation=False)

    # Use all data for training (no validation split)
    print("Using all data for training")

    # Create data loader for training
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)

    # For debugging purposes, let's check the shape of a sample
    sample_input, _ = train_dataset[0]
    print(f"Sample input shape: {sample_input.shape}")

    # Initialize model
    print("Creating HybridSN model")
    model = HybridSN(
        in_channels=data.shape[0],
        patch_size=patch_size,
        n_classes=num_classes
    ).to(device)

    # Print model summary
    print(model)
    total_params = sum(p.numel() for p in model.parameters())
    print(f"Total parameters: {total_params:,}")

    # Assuming train_dataset is already defined, you can update the class weight calculation as follows:
    labels = [train_dataset[i][1] for i in range(len(train_dataset))]

    # Convert labels to a numpy array before computing class weights
    labels = np.array(labels)

    # Compute class weights using sklearn's compute_class_weight
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)

    # Convert class_weights to a PyTorch tensor
    class_weights = torch.tensor(class_weights, dtype=torch.float32)

    # Ensure that the class_weights tensor is on the same device as your model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    class_weights = class_weights.to(device)

    # Define the criterion with the class weights
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    # Optimizer with weight decay
    optimizer = optim.Adam(model.parameters(), lr=0.0006, weight_decay=1e-4)

    # Train the model
    print("Training the model...")
    trained_model, train_losses = train_model(
        model, train_loader, criterion, optimizer, device, num_epochs=20
    )

    # Use the trained model directly for prediction
    print("Generating segmentation map using the trained model...")
    pred_map, pred_map_crf, prob_map = predict_image(trained_model, data, patch_size, num_classes, device)

    # Save the prediction maps
    np.save(f'indian_pines_prediction_map_hybridsn_patch{patch_size}.npy', pred_map)
    np.save(f'indian_pines_prediction_map_crf_patch{patch_size}.npy', pred_map_crf)

    # Visualize results
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(gt)
    plt.title('Ground Truth')
    plt.colorbar()

    plt.subplot(1, 3, 2)
    plt.imshow(pred_map)
    plt.title(f'Raw Prediction')
    plt.colorbar()

    plt.subplot(1, 3, 3)
    plt.imshow(pred_map_crf)
    plt.title(f'Post-processed Prediction')
    plt.colorbar()

    plt.tight_layout()
    plt.savefig(f'indian_pines_segmentation_results_hybridsn_patch{patch_size}.png')
    plt.show()

    print("Segmentation complete!")

    return trained_model, pred_map, pred_map_crf

if __name__ == "__main__":
    patch_size = 5  # Larger patch size for better spatial context
    run_indian_pines_segmentation(patch_size)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import scipy.io as sio
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, f1_score, classification_report
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from scipy import ndimage

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

class HybridSN(nn.Module):
    def __init__(self, in_channels, patch_size, n_classes):
        super(HybridSN, self).__init__()

        self.patch_size = patch_size

        # 3D Convolution layer
        self.conv3d_1 = nn.Conv3d(1, 8, kernel_size=(3, 3, 3), stride=(1, 1, 1), padding=(1, 1, 1))
        self.bn3d_1 = nn.BatchNorm3d(8)

        # Determine spatial dimensions after 3D operations
        dummy_input = torch.zeros(1, 1, in_channels, patch_size, patch_size)
        x = F.relu(self.bn3d_1(self.conv3d_1(dummy_input)))

        # Reshape for 2D convolution (N,C,D,H,W) -> (N,C*D,H,W)
        spatial_dim = patch_size
        n_dim = x.size(1) * x.size(2)  # Channels * Depth
        x = x.view(x.size(0), n_dim, spatial_dim, spatial_dim)

        # 2D Convolution layer
        self.conv2d_1 = nn.Conv2d(n_dim, 64, kernel_size=3, stride=1, padding=1)
        self.bn2d_1 = nn.BatchNorm2d(64)

        # Global average pooling
        x = F.relu(self.bn2d_1(self.conv2d_1(x)))
        x = F.adaptive_avg_pool2d(x, (1, 1))
        fc_size = x.view(x.size(0), -1).shape[1]

        # Fully connected layer
        self.fc1 = nn.Linear(fc_size, n_classes)

        print(f"Model initialized with FC input size: {fc_size}")

    def forward(self, x):
        # 3D Convolution path
        x = F.relu(self.bn3d_1(self.conv3d_1(x)))

        # Reshape for 2D convolution (N,C,D,H,W) -> (N,C*D,H,W)
        x = x.view(x.size(0), -1, x.size(3), x.size(4))

        # 2D Convolution path
        x = F.relu(self.bn2d_1(self.conv2d_1(x)))

        # Global average pooling
        x = F.adaptive_avg_pool2d(x, (1, 1))
        x = x.view(x.size(0), -1)

        # Fully connected path
        x = self.fc1(x)

        return x

# Custom Dataset for Indian Pines excluding background and ensuring valid patch extraction
class IndianPinesNoBackgroundDataset(Dataset):
    def __init__(self, data, labels, patch_size=11, transforms=None, use_augmentation=False):
        self.data = data
        self.labels = labels
        self.patch_size = patch_size
        self.transforms = transforms
        self.use_augmentation = use_augmentation
        self.bands, self.height, self.width = data.shape
        self.pad = patch_size // 2
        self.valid_indices = self._get_valid_indices()

    def _get_valid_indices(self):
        valid_coords = []
        for i in range(self.pad, self.height - self.pad):
            for j in range(self.pad, self.width - self.pad):
                if self.labels[i, j] != 0:
                    valid_coords.append((i, j, self.labels[i, j]))
        return valid_coords

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        i, j, label = self.valid_indices[idx]

        # Extract patch
        patch = self.data[:, i - self.pad : i + self.pad + 1, j - self.pad : j + self.pad + 1].copy()

        # Adjust label to be 0-indexed
        label = label - 1

        # Apply data augmentation (same as before)
        if self.use_augmentation and np.random.random() > 0.5:
            if np.random.random() > 0.5:
                patch = np.flip(patch, axis=1).copy()
            if np.random.random() > 0.5:
                patch = np.flip(patch, axis=2).copy()
            k = np.random.randint(1, 4)
            patch = np.rot90(patch, k=k, axes=(1, 2)).copy()
            if np.random.random() > 0.7:
                noise = np.random.normal(0, 0.01, patch.shape)
                patch = patch + noise

        # Convert to tensor
        patch = torch.FloatTensor(patch).unsqueeze(0)

        if self.transforms:
            patch = self.transforms(patch)

        return patch, label

# Function to load and preprocess the data (remains the same)
def load_indian_pines_data(data_path='Indian_pines.mat', gt_path='Indian_pines_gt.mat'):
    data = sio.loadmat(data_path)['indian_pines']
    gt = sio.loadmat(gt_path)['indian_pines_gt']
    scaler = MinMaxScaler()
    h, w, b = data.shape
    data_reshaped = data.reshape((h * w, b))
    data_normalized = scaler.fit_transform(data_reshaped)
    data = data_normalized.reshape((h, w, b))
    data = np.transpose(data, (2, 0, 1))
    return data, gt

def train_model(model, train_loader, criterion, optimizer, device, num_epochs=100):
    train_losses = []
    val_losses = []
    val_accuracies = []
    val_f1_scores = []

    best_val_acc = 0.0
    best_val_f1 = 0.0
    best_model_state = None

    patience = 15
    counter = 0

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0

        loop = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')

        for inputs, labels in loop:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            loop.set_postfix(loss=loss.item())

        epoch_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_loss)

        # Validation (using the same data as training for simplicity in this modified example)
        model.eval()
        val_loss = 0.0
        val_preds = []
        val_true = []

        with torch.no_grad():
            for inputs, labels in train_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                val_preds.extend(preds.cpu().numpy())
                val_true.extend(labels.cpu().numpy())

        val_epoch_loss = val_loss / len(train_loader.dataset)
        val_losses.append(val_epoch_loss)
        val_acc = accuracy_score(val_true, val_preds)
        val_accuracies.append(val_acc)
        val_f1 = f1_score(val_true, val_preds, average='weighted')
        val_f1_scores.append(val_f1)

        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_loss:.4f}, '
              f'Val Loss: {val_epoch_loss:.4f}, '
              f'Val Acc: {val_acc:.4f}, '
              f'Val F1: {val_f1:.4f}')

        if val_f1 > best_val_f1:
            best_val_f1 = val_f1
            best_val_acc = val_acc
            best_model_state = model.state_dict().copy()
            counter = 0
            print(f"New best model saved with F1 Score: {val_f1:.4f}")
        else:
            counter += 1

        if counter >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

    if best_model_state is not None:
        model.load_state_dict(best_model_state)
        print(f"Loaded best model with F1 Score: {best_val_f1:.4f}")

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.plot(val_accuracies, label='Val Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.plot(val_f1_scores, label='Val F1 Score')
    plt.xlabel('Epoch')
    plt.ylabel('F1 Score')
    plt.legend()

    plt.tight_layout()
    plt.savefig(f'training_curves_hybridsn_nobackground_patch{model.patch_size}.png')

    print(f"Training complete. Best validation accuracy: {best_val_acc:.4f}, Best F1 Score: {best_val_f1:.4f}")

    return model, train_losses

def predict_image(model, data, patch_size, num_classes, device):
    model.eval()
    bands, height, width = data.shape
    pad = patch_size // 2
    pred_map = np.zeros((height, width))
    padded_data = np.pad(data, ((0, 0), (pad, pad), (pad, pad)), mode='reflect')
    batch_size = 32
    coords = [(i, j) for i in range(height) for j in range(width)]
    num_batches = (len(coords) + batch_size - 1) // batch_size
    prob_map = np.zeros((height, width, num_classes))

    with torch.no_grad():
        for batch_idx in tqdm(range(num_batches), desc="Predicting"):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(coords))
            batch_coords = coords[start_idx:end_idx]
            batch_patches = []
            for i, j in batch_coords:
                patch = padded_data[:, i:i+patch_size, j:j+patch_size]
                batch_patches.append(patch)
            batch_patches = torch.FloatTensor(np.array(batch_patches)).unsqueeze(1).to(device)
            outputs = model(batch_patches)
            probs = torch.softmax(outputs, dim=1).cpu().numpy()
            _, preds = torch.max(outputs, 1)
            for idx, (i, j) in enumerate(batch_coords):
                pred_map[i, j] = preds[idx].item() + 1 # Keep 1-based indexing for output map
                prob_map[i, j] = probs[idx]

    try:
        import pydensecrf.densecrf as dcrf
        from pydensecrf.utils import unary_from_softmax
        print("Applying CRF post-processing...")
        d = dcrf.DenseCRF2D(width, height, num_classes)
        U = unary_from_softmax(np.transpose(prob_map, (2, 0, 1)))
        d.setUnaryEnergy(U)
        d.addPairwiseGaussian(sxy=3, compat=3)
        d.addPairwiseBilateral(sxy=80, srgb=13, rgbim=np.transpose(data[:3], (1, 2, 0)), compat=10)
        Q = d.inference(5)
        pred_map_crf = np.argmax(np.array(Q).reshape((num_classes, height, width)), axis=0) + 1
        print("CRF post-processing complete.")
    except ImportError:
        print("pydensecrf not available, using median filter instead.")
        from scipy import ndimage
        pred_map_crf = ndimage.median_filter(pred_map, size=3)

    return pred_map, pred_map_crf, prob_map


def evaluate_model(model, data, gt, patch_size, device):
    model.eval()
    bands, height, width = data.shape
    pad = patch_size // 2
    padded_data = np.pad(data, ((0, 0), (pad, pad), (pad, pad)), mode='reflect')
    all_preds = []
    all_labels = []
    coords = np.argwhere(gt != 0) # Evaluate only on non-background pixels
    batch_size = 32
    num_batches = (len(coords) + batch_size - 1) // batch_size

    with torch.no_grad():
        for batch_idx in tqdm(range(num_batches), desc="Evaluating"):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(coords))
            batch_coords = coords[start_idx:end_idx]
            batch_patches = []
            batch_gt_labels = []
            for i, j in batch_coords:
                patch = padded_data[:, i:i+patch_size, j:j+patch_size]
                batch_patches.append(patch)
                batch_gt_labels.append(gt[i, j] - 1) # Adjust ground truth labels to be 0-indexed
            batch_patches = torch.FloatTensor(np.array(batch_patches)).unsqueeze(1).to(device)
            outputs = model(batch_patches)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(
                batch_gt_labels)

    all_labels = np.array(all_labels)
    all_preds = np.array(all_preds)

    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    report = classification_report(all_labels, all_preds)

    print(f"Evaluation Accuracy on Non-Background Pixels: {accuracy:.4f}")
    print(f"Evaluation F1 Score on Non-Background Pixels: {f1:.4f}")
    print("Classification Report on Non-Background Pixels:\n", report)

def run_indian_pines_segmentation_no_background(patch_size=11):
    if patch_size % 2 == 0:
        print(f"Warning: Patch size {patch_size} is even. Using {patch_size+1} instead.")
        patch_size = patch_size + 1

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'Using device: {device}')

    print("Loading and preprocessing data...")
    data, gt = load_indian_pines_data()

    # Filter ground truth to get unique non-zero labels
    non_zero_labels = np.unique(gt[gt != 0])
    num_classes = len(non_zero_labels) # Should be 16 - 1 = 15

    print(f"Number of classes (excluding background): {num_classes}")
    print(f"Using patch size: {patch_size}")

    # Create dataset without background for training
    train_dataset = IndianPinesNoBackgroundDataset(data, gt, patch_size=patch_size, use_augmentation=False)
    print(f"Training dataset size: {len(train_dataset)}")

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)

    sample_input, _ = train_dataset[0]
    print(f"Sample input shape: {sample_input.shape}")

    print("Creating HybridSN model")
    model = HybridSN(
        in_channels=data.shape[0],
        patch_size=patch_size,
        n_classes=num_classes
    ).to(device)

    print(model)
    total_params = sum(p.numel() for p in model.parameters())
    print(f"Total parameters: {total_params:,}")

    # Compute class weights based on the filtered dataset
    labels = [train_dataset[i][1] for i in range(len(train_dataset))]
    labels = np.array(labels)
    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
    class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
    criterion = nn.CrossEntropyLoss(weight=class_weights)
    optimizer = optim.Adam(model.parameters(), lr=0.0006, weight_decay=1e-4)

    print("Training the model...")
    trained_model, train_losses = train_model(
        model, train_loader, criterion, optimizer, device, num_epochs=20
    )

    print("Generating segmentation map using the trained model...")
    # When predicting, we still predict for all pixels, but the model is trained to output 15 classes.
    pred_map, pred_map_crf, prob_map = predict_image(trained_model, data, patch_size, num_classes, device)

    # For visualization, mask out the background in the ground truth
    masked_gt = gt.astype(int).copy() # Cast to int to allow negative values
    masked_gt[masked_gt == 0] = -1 # Use a different value for background in visualization

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(masked_gt)
    plt.title('Ground Truth (Background Masked)')
    plt.colorbar()

    plt.subplot(1, 3, 2)
    plt.imshow(pred_map)
    plt.title(f'Raw Prediction (15 classes)')
    plt.colorbar()

    plt.subplot(1, 3, 3)
    plt.imshow(pred_map_crf)
    plt.title(f'Post-processed Prediction (15 classes)')
    plt.colorbar()

    plt.tight_layout()
    plt.savefig(f'indian_pines_segmentation_results_nobackground_patch{patch_size}.png')
    plt.show()

    # Evaluate the model on the non-background pixels
    print("Evaluating the model on non-background pixels...")
    evaluate_model(trained_model, data, gt, patch_size, device)

    print("Segmentation complete!")

    return trained_model, pred_map, pred_map_crf

if __name__ == "__main__":
    patch_size = 5
    run_indian_pines_segmentation_no_background(patch_size)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import scipy.io as sio
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, f1_score, classification_report
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from scipy import ndimage

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# (HybridSN and IndianPinesNoBackgroundDataset remain the same as in the previous response)

# Function to load and preprocess Pavia University data
def load_pavia_university_data(data_path='PaviaU.mat', gt_path='PaviaU_gt.mat'):
    data = sio.loadmat(data_path)['paviaU']
    gt = sio.loadmat(gt_path)['paviaU_gt']

    scaler = MinMaxScaler()
    h, w, b = data.shape
    data_reshaped = data.reshape((h * w, b))
    data_normalized = scaler.fit_transform(data_reshaped)
    data = data_normalized.reshape((h, w, b))
    data = np.transpose(data, (2, 0, 1))
    return data, gt

# Custom Dataset for Pavia University
class PaviaUniversityDataset(Dataset):
    def __init__(self, data, labels, patch_size=11, transforms=None):
        self.data = data
        self.labels = labels
        self.patch_size = patch_size
        self.transforms = transforms
        self.bands, self.height, self.width = data.shape
        self.pad = patch_size // 2
        self.valid_indices = self._get_valid_indices()

    def _get_valid_indices(self):
        valid_coords = []
        for i in range(self.pad, self.height - self.pad):
            for j in range(self.pad, self.width - self.pad):
                valid_coords.append((i, j))
        return valid_coords

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        i, j = self.valid_indices[idx]
        patch = self.data[:, i - self.pad : i + self.pad + 1, j - self.pad : j + self.pad + 1].copy()
        label = self.labels[i, j] - 1 # 0-indexed labels for Pavia University

        patch = torch.FloatTensor(patch).unsqueeze(0)
        if self.transforms:
            patch = self.transforms(patch)
        return patch, label

def predict_image(model, data, patch_size, num_output_classes, device):
    model.eval()
    bands_data, height, width = data.shape
    pad = patch_size // 2
    pred_map = np.zeros((height, width))
    padded_data = np.pad(data, ((0, 0), (pad, pad), (pad, pad)), mode='reflect')
    batch_size = 32
    coords = [(i, j) for i in range(height) for j in range(width)]
    num_batches = (len(coords) + batch_size - 1) // batch_size
    prob_map = np.zeros((height, width, num_output_classes))

    with torch.no_grad():
        for batch_idx in tqdm(range(num_batches), desc="Predicting PaviaU"):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(coords))
            batch_coords = coords[start_idx:end_idx]
            batch_patches = []
            for i, j in batch_coords:
                patch = padded_data[:, i:i+patch_size, j:j+patch_size]
                batch_patches.append(patch)

            # Prepare input tensor with the correct number of channels (1)
            batch_tensor = torch.FloatTensor(np.array(batch_patches)).unsqueeze(1).to(device)

            try:
                outputs = model(batch_tensor)
                probs = torch.softmax(outputs, dim=1).cpu().numpy()
                _, preds = torch.max(outputs, 1)
                for idx, (i, j) in enumerate(batch_coords):
                    pred_map[i, j] = preds[idx].item() + 1
                    prob_map[i, j] = probs[idx]
            except RuntimeError as e:
                print(f"RuntimeError during prediction: {e}")
                print(f"Input tensor shape: {batch_tensor.shape}")
                # Optionally, inspect the weights of the first conv layer
                # print(f"Conv3d_1 weight shape: {model.conv3d_1.weight.shape}")
                raise e

    try:
        import pydensecrf.densecrf as dcrf
        from pydensecrf.utils import unary_from_softmax
        print("Applying CRF post-processing...")
        d = dcrf.DenseCRF2D(width, height, num_output_classes)
        U = unary_from_softmax(np.transpose(prob_map, (2, 0, 1)))
        d.setUnaryEnergy(U)
        d.addPairwiseGaussian(sxy=3, compat=3)
        d.addPairwiseBilateral(sxy=80, srgb=13, rgbim=np.transpose(data[:3], (1, 2, 0)), compat=10)
        Q = d.inference(5)
        pred_map_crf = np.argmax(np.array(Q).reshape((num_output_classes, height, width)), axis=0) + 1
        print("CRF post-processing complete.")
    except ImportError:
        print("pydensecrf not available, using median filter instead.")
        from scipy import ndimage
        pred_map_crf = ndimage.median_filter(pred_map, size=3)

    return pred_map, pred_map_crf, prob_map

def evaluate_model_pavia(model, data, gt, patch_size, device, num_indian_pines_classes=15):
    model.eval()
    bands, height, width = data.shape
    pad = patch_size // 2
    padded_data = np.pad(data, ((0, 0), (pad, pad), (pad, pad)), mode='reflect')
    all_preds = []
    all_labels = []
    coords = [(i, j) for i in range(height) for j in range(width)]
    batch_size = 32
    num_batches = (len(coords) + batch_size - 1) // batch_size

    with torch.no_grad():
        for batch_idx in tqdm(range(num_batches), desc="Evaluating on PaviaU"):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(coords))
            batch_coords = coords[start_idx:end_idx]
            batch_patches = []
            batch_gt_labels = []
            for i, j in batch_coords:
                patch = padded_data[:, i:i+patch_size, j:j+patch_size]
                batch_patches.append(patch)
                batch_gt_labels.append(gt[i, j] - 1) # 0-indexed Pavia labels
            batch_patches = torch.FloatTensor(np.array(batch_patches)).unsqueeze(1).to(device)
            outputs = model(batch_patches)
            _, preds = torch.max(outputs, 1) # Get the most likely of the 15 trained classes
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(np.array(batch_gt_labels))

    all_labels = np.array(all_labels)
    all_preds = np.array(all_preds)

    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted', labels=np.unique(all_labels))
    report = classification_report(all_labels, all_preds, labels=np.unique(all_labels))

    print(f"Evaluation Accuracy on Pavia University (using Indian Pines trained model): {accuracy:.4f}")
    print(f"Evaluation F1 Score on Pavia University (using Indian Pines trained model): {f1:.4f}")
    print("Classification Report on Pavia University (using Indian Pines trained model):\n", report)

def run_cross_dataset_evaluation(indian_pines_patch_size=5, pavia_patch_size=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'Using device: {device}')

    print("Loading and preprocessing Indian Pines data for training...")
    data_ip, gt_ip = load_indian_pines_data()

    # Filter ground truth for Indian Pines to get unique non-zero labels for training
    non_zero_labels_ip_train = np.unique(gt_ip[gt_ip != 0])
    num_classes_ip_train = len(non_zero_labels_ip_train)
    print(f"Number of classes (excluding background) for Indian Pines training: {num_classes_ip_train}")

    # Create dataset without background for Indian Pines training
    train_dataset_ip = IndianPinesNoBackgroundDataset(data_ip, gt_ip, patch_size=indian_pines_patch_size, use_augmentation=False)
    train_loader_ip = DataLoader(train_dataset_ip, batch_size=32, shuffle=True, num_workers=4)

    print("Creating HybridSN model trained on Indian Pines...")
    model_ip = HybridSN(
        in_channels=data_ip.shape[0],
        patch_size=indian_pines_patch_size,
        n_classes=num_classes_ip_train
    ).to(device)

    # Compute class weights for Indian Pines
    labels_ip_train = [train_dataset_ip[i][1] for i in range(len(train_dataset_ip))]
    labels_ip_train = np.array(labels_ip_train)
    class_weights_ip = compute_class_weight('balanced', classes=np.unique(labels_ip_train), y=labels_ip_train)
    class_weights_ip = torch.tensor(class_weights_ip, dtype=torch.float32).to(device)
    criterion_ip = nn.CrossEntropyLoss(weight=class_weights_ip)
    optimizer_ip = optim.Adam(model_ip.parameters(), lr=0.0006, weight_decay=1e-4)

    print("Training the model on Indian Pines (without background)...")
    trained_model_ip, train_losses_ip = train_model(
        model_ip, train_loader_ip, criterion_ip, optimizer_ip, device, num_epochs=20
    )

    print("\nLoading and preprocessing Pavia University data for testing...")
    data_pu, gt_pu = load_pavia_university_data()
    num_classes_pu = np.max(gt_pu)
    print(f"Number of classes in Pavia University: {num_classes_pu}")
    in_channels_pu = data_pu.shape[0]

    # Modify the first 2D conv layer to accept the new number of input channels
    original_conv2d_weight = trained_model_ip.conv2d_1.weight.data
    original_out_channels = trained_model_ip.conv2d_1.out_channels
    original_kernel_size = trained_model_ip.conv2d_1.kernel_size
    original_stride = trained_model_ip.conv2d_1.stride
    original_padding = trained_model_ip.conv2d_1.padding
    original_bias = trained_model_ip.conv2d_1.bias

    # Create a new Conv2d layer with the correct input channels
    new_conv2d_1 = nn.Conv2d(8 * in_channels_pu, original_out_channels, original_kernel_size, original_stride, original_padding)

    # Try to initialize the weights of the new layer (can be simple random initialization)
    nn.init.kaiming_normal_(new_conv2d_1.weight, mode='fan_out', nonlinearity='relu')
    if original_bias is not None:
        new_conv2d_1.bias = nn.Parameter(torch.zeros_like(original_bias))

    # Replace the old conv2d_1 layer with the new one in the loaded model
    trained_model_ip.conv2d_1 = new_conv2d_1.to(device)

    print("Generating segmentation map for Pavia University using the (modified) Indian Pines trained model...")
    pred_map_pu, pred_map_crf_pu, prob_map_pu = predict_image(trained_model_ip, data_pu, pavia_patch_size, num_classes_ip_train, device)

    # Visualize predictions on Pavia University
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(gt_pu)
    plt.title('Pavia University Ground Truth')
    plt.colorbar()

    plt.subplot(1, 3, 2)
    plt.imshow(pred_map_pu)
    plt.title(f'Prediction on PaviaU (Modified IP Model)')
    plt.colorbar()

    plt.subplot(1, 3, 3)
    plt.imshow(pred_map_crf_pu)
    plt.title(f'Prediction on PaviaU (Modified IP Model, CRF)')
    plt.colorbar()

    plt.tight_layout()
    plt.savefig(f'pavia_prediction_using_modified_ip_model_patch{pavia_patch_size}.png')
    plt.show()

    # Evaluate on Pavia University (results will likely be poor)
    print("\nEvaluating the (modified) Indian Pines trained model on Pavia University...")
    evaluate_model_pavia(trained_model_ip, data_pu, gt_pu, pavia_patch_size, device, num_classes_ip_train)

if __name__ == "__main__":
    indian_pines_patch_size = 5
    pavia_university_patch_size = 5
    run_cross_dataset_evaluation(indian_pines_patch_size, pavia_university_patch_size)
```
